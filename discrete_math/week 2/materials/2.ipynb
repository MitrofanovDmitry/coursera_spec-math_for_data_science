# Тест 2

Помечаем категорию для каждой видеозаписи

```python
counter = 0
for c in combinations_with_replacement(range(4), 15):
    counter += 1
    print(c)
counter
```

```python
from itertools import combinations

counter = 0
for c in combinations(range(18), 3):
    counter += 1
    print(c)
counter
```
