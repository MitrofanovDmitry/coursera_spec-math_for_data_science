```python
import networkx as nx
import random


def Previsit(v):
    global clock
    pre[v] = clock
    clock += 1

def Postvisit(v):
    global clock
    post[v] = clock
    clock += 1

def Explore(v):
    visited[v] = True
    Previsit(v)
    for u in graph[v]:
        if not visited[u]:
            Explore(u)
    Postvisit(v)

def dfs():
    for v in graph:
        if not visited[v]:
            Explore(v)

def vertex_deletion():
    dfs()
    
    delete = []
    for key, value in post.items():
        if post[key] - pre[key] == 1:
            delete.append(key)
            
    vertex = random.choice(delete)

    return vertex
```

```python
graph = nx.Graph()
graph.add_nodes_from(["A", "B", "C", "D", "E"])
graph.add_edges_from([("A", "B"), ("B", "C"), ("C", "A"), ("B", "D"), ("B", "E"), ("D", "E")])

clock = 0
pre = {}
post = {}

visited = {}
for v in graph:
    visited[v] = False

print(vertex_deletion())
```
