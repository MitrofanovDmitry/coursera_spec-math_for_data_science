# Ориентированные графы

В этом ноутбуке мы посмотрим на примеры работы с ориентированными графами

```python
# Задаем граф

import networkx as nx

graph = nx.DiGraph()

graph.add_edge("A", "B")
graph.add_edges_from([("B", "C"), ("C", "A"), ("C", "D"), ("C", "E"), ("D", "E")])

graph.add_node("F")

nodes = list(graph.nodes())
edges = list(graph.edges())

print(f"Список вершин: {nodes}")
print(f"Список ребер: {edges}")

num_nodes = graph.number_of_nodes()
num_edges = graph.number_of_edges()

print(f"Число вершин: {num_nodes}")
print(f"Число ребер: {num_edges}")
```

```python
# Риусем граф

%matplotlib inline

nx.draw_networkx(graph)
```

```python
# Рисуем граф примерно как в слайдах

import matplotlib.pyplot as plt

pos = {
    "A": (1, 1.5),
    "B": (4, 0),
    "C": (5, 2),
    "D": (2.9, 2.2),
    "E": (6, 0.5),
    "F": (7.5, 1.5),
}


def draw_graph():
    options = {
        "font_size": 20,
        "node_size": 1500,
        "node_color": "white",
        "edgecolors": "black",
        "linewidths": 3,
        "width": 3
    }
    plt.axis([0, 8.5, -0.5, 2.7])
    nx.draw_networkx(graph, pos, **options)
    plt.show()


draw_graph()
```

```python
# Ищем степень вершины

print(f"Степень вершины: {graph.degree('C')}")
print(f"Входящая степень вершины: {graph.in_degree('C')}")
print(f"Исходящая степень вершины: {graph.out_degree('C')}")
```

```python
# Ищем соседей

print(f"Что выдает graph[v]: {list(graph['C'])}")
print(f"А вот соседи по входящим ребрам: {list(graph.pred['C'])}")
```

```python
# Проверяем достижимость

nx.has_path(graph, "D", "A")
```

```python
# Проверяем связность

nx.is_strongly_connected(graph)
```
