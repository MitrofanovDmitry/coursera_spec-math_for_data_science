```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from numpy.linalg import svd
```

```python
wine = pd.read_csv("Wine.csv")
```

### 0. Посмотрим на данные

```python
wine.describe().T
```

```python
sns.jointplot(data=wine, x="Alcohol", y="Malic_Acid", hue="Customer_Segment");
```

# 1. SVD вручную

```python
wine_wo_segment = wine.drop(columns="Customer_Segment")
```

```python
scaler = StandardScaler()
wine_scaled = scaler.fit_transform(wine_wo_segment)
```

```python
wine_scaled[:3, :]
```

```python
U, S, Vt = svd(wine_scaled)
```

```python
Sigma = np.zeros(wine_scaled.shape)
Sigma[0, 0] = S[0]
Sigma[1, 1] = S[1]
```

```python
wine_pc = U @ Sigma
pc_1 = wine_pc[:, 0]
pc_2 = wine_pc[:, 1]

pc_1[:5]
```

```python
S**2 / np.sum(S**2)
```

# 2. PCA

```python
pca = PCA(n_components=2)
```

```python
wine_pca = pca.fit_transform(wine_scaled)
```

```python
wine["pc_1"] = wine_pca[:, 0]
wine["pc_2"] = wine_pca[:, 1]
```

```python
wine["pc_1"][:5]
```

```python
sns.jointplot(data=wine, x="pc_1", y="pc_2", hue="Customer_Segment")
```

```python
pca.explained_variance_ratio_
```
